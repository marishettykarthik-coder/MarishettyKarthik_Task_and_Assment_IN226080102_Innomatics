```python
#PROBLEM 1:UNIQUE WORDS IN A SENTENCE

user_input = input("Enter a sentence: ")    
#convert the sentence to lowercase and split it into words
words = user_input.lower().split()
unique_words = set(words) 
print("Unique words count:",len(unique_words))
print("Unique words:",unique_words)
```

```python
#problem 2: HIGHEST SALARY FROM EMPLOYEE DATA
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
highest_salary = 0
for amount in employees.values():
    if amount > highest_salary:
        highest_salary = amount
print("Highest salary:", highest_salary, "is earned by", [name for name, salary in employees.items() if salary == highest_salary][0])
```

```python
#Problem 3: Find Maximum and Minimum Values

numbers = [45, 22, 89, 10, 66]

# Initialize max and min with first element
max_value = numbers[0]
min_value = numbers[0]

# Loop through the list starting from second element
for number in numbers[1:]:
    if number > max_value:
        max_value = number
    if number < min_value:
        min_value = number

# final output
print("List:", numbers)
print("Max:", max_value)
print("Min:", min_value)
```

```python
#Problem 4: Count Products Above a Price Threshold

prices = [450, 1200, 899, 1500, 300]

amount = 1000
count = 0

for price in prices:
    if price > amount:
        count += 1

print("Products above", amount, "are:", count)
```

```python
# Problem 5: Calculate Attendance Percentage

attendance = ["P", "P", "A", "P", "P"]

# Counting total days
total_days = len(attendance)

# Counting number of presents
present_count = 0

for i in attendance:
    if i == "P":
        present_count += 1

# Calculating percentage
attendance_percentage = (present_count / total_days) * 100

# final output
print("Attendance Percentage:", attendance_percentage)
```

```python
# Problem 6: Remove Duplicate Phone Numbers

phone_numbers = [9876543210, 9123456789, 9876543210]

# Converting list to set to remove duplicates
unique_numbers = set(phone_numbers)

# final output
print("Unique phone numbers:", unique_numbers)
```

```python
#Problem 7: Count Character Frequency

text = input("Enter a string: ")

char_freq = {}

for ch in text:
    if ch in char_freq:
        char_freq[ch] += 1
    else:
        char_freq[ch] = 1

print("Character Frequency:", char_freq)
```
