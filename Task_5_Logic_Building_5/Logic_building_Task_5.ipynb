```python
# Problem 1: Social Media - Post Engagement Analyzer

def analyze_engagement(likes_list):
    total_likes = 0
    for like in likes_list:
        total_likes += like

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)

likes = [200, 300, 150, 400]
analyze_engagement(likes)

```

```python
# Problem 2: Healthcare - Medicine Stock Alert System

def check_medicine_stock(stock):
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    print("Medicine Stock:", stock)
    print("Status:", status)

check_medicine_stock(6)

```

```python
# Problem 3: Agriculture - Rainfall Adequacy Checker

def check_rainfall(rainfall_list, required_level):
    total = 0
    for rain in rainfall_list:
        total += rain

    average = total / len(rainfall_list)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", average)
    print("Rainfall Status:", status)

rainfall_data = [70, 71, 90, 65, 70]
check_rainfall(rainfall_data, 70)

```

```python
# Problem 4: Social Media - Duplicate Account Detection

def detect_duplicates(usernames):
    unique_users = set(usernames)

    if len(unique_users) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

users = ["user1", "user2", "user3", "user1"]
detect_duplicates(users)

```

```python
# Problem 5: Healthcare - Appointment Eligibility Checker

def check_appointment_eligibility(age):
    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Patient Age:", age)
    print("Eligibility Status:", status)

check_appointment_eligibility(21)

```

```python
# Problem 6: Agriculture - Premium Crop Price Filter

def filter_premium_crops(prices):
    premium = []
    for price in prices:
        if price > 2000:
            premium.append(price)

    print("Premium Crops:", premium)

crop_prices = [1500, 2500, 1800, 3200]
filter_premium_crops(crop_prices)

```

```python
# Problem 7: System Monitoring - Application Health Checker

def check_system_health(error_count):
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", error_count)
    print("System Status:", status)

check_system_health(7)

```

```python
# Problem 8: Banking - Daily Transaction Limit Checker

def check_transaction(amount):
    daily_limit = 50000

    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"

    print("Transaction Amount:", amount)
    print("Transaction Status:", status)

check_transaction(60000)

```

```python
# Problem 9: E-Learning - Student Attendance Eligibility Syste

def check_exam_eligibility(attendance_list):
    present = 0
    for day in attendance_list:
        if day == "P":
            present += 1

    percentage = (present / len(attendance_list)) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)

attendance = ["P", "P", "A", "P", "P"]
check_exam_eligibility(attendance)

```

```python
# Problem 10: Smart Electricity Bill Analyzer

def calculate_electricity_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage

total_bill, usage_status = calculate_electricity_bill(250)

print("Total Bill:", total_bill)
print("Usage Status:", usage_status)


```
