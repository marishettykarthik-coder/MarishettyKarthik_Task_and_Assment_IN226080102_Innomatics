```python
# Problem 1: Mobile Recharge Validation System

def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: rupees"))

        if amount < 50:
            print("Recharge amount must be at least 50 rupees.")
        elif amount not in valid_plans:
            print("Enter amount",amount)
            print("Invalid plan. Available plans:", valid_plans)
        else:
            print("Enter amount",amount)
            print("Recharge Successful!")
            break

validate_recharge()

```

```python
# Problem 2: Inventory Reorder Alert System

def inventory_check(products):
    for product, stock in products.items():
        if stock < 15:
            print(product, ": Reorder Alert")
        else:
            print(product, ": Stock OK")

products = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 12,
    "Monitor": 18
}

inventory_check(products)

```

```python
# Problem 3: Student Result Processing System


def process_result(marks):
    total = 0

    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"

student_marks = [70, 65, 43, 55]

result = process_result(student_marks)
print("average", sum(student_marks) / len(student_marks))
print("Result:", result)

```

```python
# Problem 4: Cab Fare Estimator with Retry Option

def calculate_fare(distance, peak_hour):
    base_fare = 50
    fare = base_fare + (12 * distance)

    if peak_hour.lower() == "yes":
        fare = fare + (fare * 0.25)

    return fare


while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ")

    total_fare = calculate_fare(distance, peak)
    print("Distance:", distance, "km")
    print("Is it peak hour?",peak)
    print("Total Fare: ₹", total_fare)

    retry = input("Do you want to calculate again? (yes/no): ")
    if retry.lower() != "yes":
        break

```

```python
# Problem 5: Employee Attendance Eligibility Checker

def check_attendance(attendance_list):
    present_days = 0

    for day in attendance_list:
        if day == "P":
            present_days += 1

    percentage = (present_days / len(attendance_list)) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]

status = check_attendance(attendance) 
print("Attendance Status:", status)

```

```python
# Problem 6: Password Strength Checker

def check_password_strength(password):
    check_digit = False
    check_special = False
    special_chars = "@#$"

    if len(password) < 8:
        return "Weak Password (Minimum length 8 required)"

    for char in password:
        if char.isdigit():
            check_digit = True
        if char in special_chars:
            check_special = True

    if check_digit and check_special:
        return "Strong Password"
    else:
        return "Weak Password (Must include digit and special character)"

password_input=input("Enter password: ")
strength=check_password_strength(password_input)
print(password_input)
print(strength)


```
