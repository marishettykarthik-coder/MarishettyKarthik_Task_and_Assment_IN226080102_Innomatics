```python
# Problem 1: Smart Parking Lot Management System

def manage_parking(capacity, logs):
    current = 0
    peak = 0

    for log in logs:
        if log == "IN":
            current += 1
        elif log == "OUT" and current > 0:
            current -= 1

        if current > peak:
            peak = current

    print("Currently Parked Vehicles:", current)

    if current > capacity:
        print("Parking Status: Exceeded Capacity")
    else:
        print("Parking Status: Available")

    print("Peak Parking Usage:", peak)


capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
manage_parking(capacity, vehicle_logs)

```

```python
# Problem 2: Online Food Delivery Time Estimator

def estimate_delivery_time(distance, traffic, weather):
    base_time = distance * 5   # 5 minutes per km baseline

    if traffic == "High":
        base_time += 15
    elif traffic == "Medium":
        base_time += 8

    if weather == "Rainy":
        base_time += 10
    elif weather == "Stormy":
        base_time += 20

    print("Estimated Delivery Time:", base_time, "minutes")


estimate_delivery_time(8, "High", "Rainy")

```

```python
# Problem 3: Movie Theatre Seat Occupancy Analyzer

def analyze_occupancy(total_seats, booked_seats):
    booked_count = 0

    for seat in booked_seats:
        if seat == 1:
            booked_count += 1

    occupancy = (booked_count / total_seats) * 100

    print("Occupancy:", int(occupancy), "%")

    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
        print("Suggestion: Consider opening additional show")
    else:
        print("Show Status: Seats Available")


total_seats = 200
booked = [1] * 150   # 150 booked seats
analyze_occupancy(total_seats, booked)

```

```python
# Problem 4: Cloud Server Load Classification System

def classify_server_load(cpu_readings):
    total = 0

    for value in cpu_readings:
        total += value

    average = total / len(cpu_readings)

    print("Average CPU Load:", int(average), "%")

    if average < 50:
        status = "Normal"
    elif average <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Server Status:", status)


cpu_data = [45, 60, 70, 85, 90]
classify_server_load(cpu_data)

```

```python
# Problem 5: Smart Classroom Resource Usage Monitor

def monitor_resources(resource_usage):
    overused = []

    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)

    if len(overused) > 0:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


usage_data = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_resources(usage_data)

```

```python
# Problem 6: Online Event Registration Capacity Controller

def manage_event_registration(capacity, registrations):
    if registrations <= capacity:
        confirmed = registrations
        waitlist = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlist = registrations - capacity
        status = "Closed"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)


manage_event_registration(100, 105)
```
