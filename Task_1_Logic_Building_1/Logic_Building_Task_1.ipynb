# 1. User Login Check

Problem Statement:
Given a username and password, check whether login is successful.

```python
username = "admin"
password = "1234"

# Taking user input
input_username = input("Enter username: ")
input_password = input("Enter password: ")

# Checking login
if input_username == username and input_password == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

# 2. Pass / Fail Analyzer
Problem Statement:
Analyze student results.

```python
# Student marks list
marks = [45, 78, 90, 33, 60]

# Initializing counters
pass_count = 0
fail_count = 0

# Iterating through each student's marks
for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Displaying final results
print("Total Students:", len(marks))
print("Total Pass Students:", pass_count)
print("Total Fail Students:", fail_count)
```

# 3. Simple Data Cleaner
Problem Statement:
Clean and standardize user names.

```python
# input data
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

# Cleaning each name
for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)
print("Cleaned Names:", cleaned_names)
```

# 4. Message Length Analyzer
Problem Statement:
Analyze message sizes.

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    print(f"Message: '{message}' | Length: {len(message)}")
    
    # Flagging long messages
    if len(message) > 10:
        print("This message is longer than 10 characters")
```

# 5. Error Message Detector
Problem Statement:
Detect error messages from system logs.


```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

# Checking each log entry
for log in logs:
    if log == "ERROR":
        error_count += 1

# Display final error count
print("Total ERROR messages:", error_count)

```
