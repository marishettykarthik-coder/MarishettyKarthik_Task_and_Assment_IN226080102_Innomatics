```python
# Problem 1: Employee Performance Bonus Eligibility


employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

highest_score = max(employees.values())

top_performers = []
for name in employees:
    if employees[name] == highest_score:
        top_performers.append(name)

print("Top Performers Eligible for Bonus:")
print(", ".join(top_performers) + " (Score: " + str(highest_score) + ")")
```

```python

# Problem 2: Search Query Keyword Analysis


import string

query = "Buy mobile phone buy phone online"

query = query.lower()
query = query.translate(str.maketrans("", "", string.punctuation))

words = query.split()

frequency = {}

for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

print("Keywords searched more than once:")
for word in frequency:
    if frequency[word] > 1:
        print(word + " : " + str(frequency[word]))
```

```python

# Problem 3: Sensor Data Validation


sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for index in range(len(sensor_readings)):
    if sensor_readings[index] % 2 == 0:
        valid_readings.append((index, sensor_readings[index]))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python

# Problem 4: Email Domain Usage Analysis


emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

for email in emails:
    domain = email.split("@")[1]
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

total_users = len(emails)

for domain in domain_count:
    percentage = (domain_count[domain] / total_users) * 100
    print(domain + " : " + str(int(percentage)) + "%")
```

```python

# Problem 5: Sales Spike Detection


sales = [1200, 1500, 900, 2200, 1400, 3000]

average_sales = sum(sales) / len(sales)
threshold = average_sales * 1.30

for i in range(len(sales)):
    if sales[i] > threshold:
        print("Day " + str(i + 1) + " : " + str(sales[i]))
```

```python

# Problem 6: Duplicate User ID Detection


user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

user_count = {}

for user in user_ids:
    if user in user_count:
        user_count[user] += 1
    else:
        user_count[user] = 1

for user in user_count:
    if user_count[user] > 1:
        print(user + " → " + str(user_count[user]) + " times")
```
